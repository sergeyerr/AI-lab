```python
gender_cats ={"Девочка" : 0, "Мальчик" : 1}
child_harm_cats = {"Не били" : 0, "Разве что по делу" : 0.3, "Часто" : 1}
second_ch_cats = {"Не знаю" : 0, "Паблик в вк" : 0.2, "Лучшая борда рунета" : 0.7, 'Ненавижу Абу' : 1}
orit_cats = {"Гетеро" : 0, "Би" : 0.5, "Альтернативная" : 1}
activity_cats = {"Сычевать" : 0, "Гулять" : 1}
jews_cats = {"Нет" : 0, "Я просто боюсь евреев" : 0.5 , "Да": 1}
crimea_cats = {"Мне всё равно" : 0, "Отрицательно" : 0.5, "Положительно": 1}
eng_cats = {'Оригатоё гайзаймас' : 0, 'Чуть лучше, чем ничего' : 0.2,
           'Нормально' : 0.4, 'Хорошо' : 0.6, 'Отлично' : 0.8, 'НЭЙТИВ ИНГЛИШ СПИКЕР' : 1}
past_cats = {'Нет' : 0,'А колбаса какая, колбаса...' : 1}
```

```python
cats_dict = {'Мальчик или девочка?' : gender_cats, 'Как часто тебя били в детстве?' : child_harm_cats,
            'Что такое два.ч?' : second_ch_cats, 'Твоя ориентация?' : orit_cats,
            'Погулять вечером после тяжкого рабочего дня, или посычевать дома?' : activity_cats,
            'Вы таки антисемит? ' : jews_cats, 'Как вы относитесь к аннексии Крыма?' : crimea_cats,
            'Как хорошо вы знаете английский?' : eng_cats, 'Было ли раньше лучше? ' : past_cats}
```

```python
def make_cats(df, res):
    for col, cats in cats_dict.items():
        res[col] = df[col].map(cats)
```

```python
films = ['Боевик','Детектив' , 'Аниме', 'Драма', 'Комедия','Мелодрама', 'Исторический фильм', 'Трагедия', 'Фильм ужасов','Фантастика']
def add_films(df, res):
    for film in films:
        res[film] = df['А какие жанры фильмов любишь? '].map(lambda x : film in x)
```

```python
science = ['Литература', 'История', 'Иностранные языки', 'Физика', 'Биология', 'Химия', 'Матеша', 'Информатика']
def add_science(df, res):
    for sci in science:
        res[sci] = df['Какой предмет ближе всего по душе?'].map(lambda x : sci == x)
```

```python
def add_originality(df, res):
       res['Оригинальность'] = df['Владимир Путин - '].map(lambda x : 'Политик, лидер и боец' != x and 'Молодец' != x) 
```

```python
num_vals = [('Оцените ваш уровень вегетарианца по 5-ти бальной шкале', 'Веганство'), ('Как часто играешь в игры?', 'Игры'),
             ('Сериалы?', 'Сериалы'), ('А книги? ', 'Книги')]
def add_num_vals(df, res):
    for f,s in num_vals:
       # f, s = *t
        res[s] = df[f] / max(df[f])
```

```python
def preprocess(df, res):
    make_cats(df, res)
    add_science(df, res)
    add_num_vals(df, res)
    add_originality(df, res)
    #res['Отношение к ВВ'] = df['Владимир Путин - ']
    add_films(df, res)
    
```

```python
res = pd.DataFrame()
preprocess(input, res)
```
